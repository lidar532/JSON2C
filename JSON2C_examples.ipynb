```python
#@title Example 1.
tester = {"name":"test1",
"source"  : 
r'''
#include <stdio.h>
/*
{ 
  "format_name_":"my_name",
  "myJSON": {
    "dest": "%s",
    "sss": "%0.2f",
    "flash": [
      "%f",
      "%4.2lf",
      "%ld",
      "%lu",
      "%u",
      "%d",
      "%c"
    ]
  }
}
 */

const char my_name[] = ""
""
    "{ "
    "\"myJSON\": {"
    "\"dest\": \"%s\","
    "\"sss\": %0.2f,"
    "\"flash\": ["
    "%f,"
    "%4.2lf,"
    "%ld,"
    "%lu,"
    "%u,"
    "%d,"
    "\"%c\""
    "]"
    "}"
    "}"
    "";

int main()
{
  char buf[128];
  int n;

    char                s[] ="a string";
    float               a   = 1.0;
    double              b   = 2.0;
    float               c   = 3.0;
    long int            d   = 4;
    unsigned long int   e   = 5;
    unsigned int        f   = 6;
    int                 g   = 7;
    char                h   = 'C';

    n=sprintf(buf,  my_name, s, a, b, c, d, e, f, g, h );
    printf("\n%d char JSON string written to char buf[].\n\n%s\n",n, buf);
    return 0;
}
'''
}

```

```python
#@title Example 2
tester = {"name":"test1",
"source"  : 
r'''
#include <stdio.h>
/*  
  {
    "format_name_" : "JSON_gps_status_fmt",
     "ppksyslogger": {
       "sss" : "%f",
       "gps_stat" : {
         "utc" : "%09.2f",
         "nsat" : "%d",
         "Q" : "%d",
         "GPS_good" : "%d",
         "raw_b" : "[%ld,%ld]",
         "ascii_b" : "[%ld,%ld]",
         "pass_b" : "[%ld,%ld]"
       }
     }
   } 
 */

/* This was autogenerated from the above JSON using js2c.ipynb */
const char JSON_gps_status_fmt[] = ""
""
    "{"
    "\"ppksyslogger\": {"
    "\"sss\" : %f,"
    "\"gps_stat\" : {"
    "\"utc\" : %09.2f,"
    "\"nsat\" : %d,"
    "\"Q\" : %d,"
    "\"GPS_good\" : %d,"
    "\"raw_b\" : \"[%ld,%ld]\","
    "\"ascii_b\" : \"[%ld,%ld]\","
    "\"pass_b\" : \"[%ld,%ld]\""
    "}"
    "}"
    "} "
    "";

int main() {
  char json_s[256];
  int n;
  float sss=1234.56, utc=012345.678;
  int nsat=3, q=4, gps_good=5;
  long int rb=6, rd=7, ab=8,ad=9, pb=10, pd=11;
  n = sprintf( json_s, JSON_gps_status_fmt, 
      sss,
      utc,
      nsat,
      q,
      gps_good,
      rb, rd,
      ab, ad,
      pb, pd
    );
    printf("\n%s\n", JSON_gps_status_fmt);
    printf("\n%d char JSON string written to char json_s[].\n\n%s\n",n, json_s);
    return 0;
}
    '''
}

```
